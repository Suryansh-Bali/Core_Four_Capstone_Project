## Code

```python
#dessert names
def dessert(url):
    soup = BeautifulSoup(requests.get(url).content, 'html.parser')
    dessert_name = soup.find('h1').text.strip()
    dessert_name=dessert_name.replace('"','')
    return dessert_name
```

```python
#dessert ingredients
def ingredients(url):
    soup = BeautifulSoup(requests.get(url).content, 'html.parser')
    try:
        ingredients = []
        for para in soup.find_all(class_='ingredients-section'):
            ingredients.append(para.get_text().strip())
        return ingredients
    except:
        return np.NaN
```

```python
#method
def instructions(url):
    soup = BeautifulSoup(requests.get(url).content, 'html.parser')
    try:
        instructions = []
        for para in soup.find_all(class_="instructions-section"):
            instructions.append(para.get_text().strip())
        return instructions
    except:
        return np.NaN
```

```python
#description, serves, cooking time, difficulty
def details(url):
    soup = BeautifulSoup(requests.get(url).content, 'html.parser')
    details = []
    for para in soup.find_all(class_ = 'recipe-meta-item'):
        details.append(para.get_text().strip())
    return details
```

```python
#dessert image
def image(url):
    soup = BeautifulSoup(requests.get(url).content, 'html.parser')
    try:
        urls = pd.Series([a.get("src") for a in soup.find_all("img")])
        urls = urls[(urls.str.contains(".jpg")==True) & 
               (urls.str.contains("imagesvc.meredithcorp.io/v3/mm/")==True) ].unique()
        pic = urls[3]
        return pic
    except:
        return np.NaN
```

```python
title = []
ing_name = []
method = []
det = []
img = []
for i in range(0,len(df['recipe_urls'])): 
    main_url = df['recipe_urls'][i]
    title.append(dessert(main_url))
    ing_name.append(ingredients(main_url))
    method.append(instructions(main_url))
    det.append(details(main_url))
    img.append(image(main_url))
    img_data = requests.get(img[i]).content
    with open(str(title[i])+".jpg", "wb") as handler:
        handler.write(img_data)
```

```python
attribs = ['recipe_name', 'details', 'ingredients','instructions','image_url','recipe_url']
new_df = pd.DataFrame(columns=attribs)
new_df['recipe_name'] = title
new_df['details'] = det
new_df['ingredients'] = ing_name
new_df['instructions'] = method
new_df['image_url'] = img
new_df['recipe_url']= df['recipe_urls']
```
