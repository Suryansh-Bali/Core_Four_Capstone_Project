```python
#dessert names
def dessert(url):
    soup = BeautifulSoup(url,"lxml")
    dessert_name = soup.find('h1').text.strip()
    dessert_name=dessert_name.replace('"','')
    return dessert_name
```

```python
#dessert ingredients
def ingredients(url):
    soup = BeautifulSoup(url,"lxml")
    ingredList = []
    if soup.find("div", {"class": "tasty-recipes-ingredients-body"}) != None:
        ingredient_container = soup.find("div", {"class": "tasty-recipes-ingredients-body"})
        if ingredient_container != None:
            for ingredient in ingredient_container.findAll('li'):
                ingredList.append(ingredient.text.replace("\xa0",'').replace("*",''))
    return ingredList
```

```python
#method
def instructions(url):
    soup = BeautifulSoup(url,"lxml")
    instructions = []
    if soup.find("div", {"class": "tasty-recipes-instructions"}) != None:
        instructions_container = soup.find("div", {"class": "tasty-recipes-instructions"}).ol
        if instructions_container == None:
            instructions_container = soup.find("div", {"class": "tasty-recipes-instructions"}).ul
        if instructions_container != None:
            i = 1
            for step in instructions_container.findAll('li'):
                instructions.append(f"Step {i}:"+step.text.replace("\xa0",' '))
                i += 1
    return instructions
```

```python
#description, serves, cooking time, difficulty
def details(url):
    soup = BeautifulSoup(url,"lxml")
    detailsLst = []
    if soup.find("div", {"class": "tasty-recipes-details"}) != None:
        details_container = soup.find("div", {"class": "tasty-recipes-details"}).ul
        if details_container != None:
            for detail in details_container.findAll('li'):
                detailsLst.append(detail.text.replace("\t",'').replace("\n",''))
        else:
            details_container = soup.find("div", {"class": "tasty-recipes-details"}).p
            detailsLst = details_container.text.split("\n")
    return detailsLst
```

```python
def description(url):
    soup = BeautifulSoup(url,"lxml")
    description = []
    try:
        for para in soup.find_all(class_ = 'tasty-recipes-description'):
            description.append(para.get_text().replace("Description",'').replace("\n",''))
            return description
    except:
        return np.NaN
        
```

```python
#dessert image
def image(url):
    soup = BeautifulSoup(url,"lxml")
    try:
        urls = pd.Series([a.get("src") for a in soup.find_all("img")])
        urls = urls[(urls.str.contains(".jpg")==True) & 
            (urls.str.contains("cdn.sallysbakingaddiction.com/wp-content/uploads/")==True)].unique()
        pic = urls[0]
        return pic
    except:
        return np.NaN
```

```python
title = []
ing_name = []
method = []
det = []
img = []
des = []
for i in range(0,len(final_df['recipe_urls'])): 
    main_url = final_df['recipe_urls'][i]
    req = Request(main_url, headers={'User-Agent': 'Mozilla/5.0'})
    web_byte = urlopen(req).read()
    webpage = web_byte.decode('utf-8')
    title.append(dessert(webpage))
    des.append(description(webpage))
    ing_name.append(ingredients(webpage))
    method.append(instructions(webpage))
    det.append(details(webpage))
    img.append(image(webpage))
```

```python
attribs = ['recipe_name','description','details', 'ingredients','instructions','image_url','recipe_url']
new_df = pd.DataFrame(columns=attribs)
new_df['recipe_name'] = title
new_df['description'] = des
new_df['details'] = det
new_df['ingredients'] = ing_name
new_df['instructions'] = method
new_df['image_url'] = img
new_df['recipe_url']= final_df['recipe_urls']
```
